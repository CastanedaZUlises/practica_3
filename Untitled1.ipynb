```python
import matplotlib.pyplot as plt

def interpolacion_Neville(x, y, target_x):
    n = len(x)
    Q = [[0] * n for _ in range(n)]

    for i in range(n):
        Q[i][0] = y[i]

    for i in range(1, n):
        for j in range(1, i + 1):
            Q[i][j] = ((target_x - x[i - j]) * Q[i][j - 1] - (target_x - x[i]) * Q[i - 1][j - 1]) / (x[i] - x[i - j])

    return Q[n - 1][n - 1]

# Ejemplo de uso
x = [3, 5, 8, 12]  # Valores conocidos de x
y = [2, 3, 5, 7]  # Valores conocidos de f(x)
target_x = 6.5  # Valor objetivo de x para interpolar f(x)

result = interpolacion_Neville(x, y, target_x)
print("Resultado:", result)

# Graficar los puntos y la funci√≥n interpolada
plt.scatter(x, y, color='red', label='Puntos conocidos')
plt.plot(target_x, result, marker='o', markersize=8, color='blue', label='Valor interpolado')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
